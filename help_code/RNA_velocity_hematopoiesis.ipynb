```python
import scvelo as scv
import scanpy as sc
import pandas as pd
import numpy as np
import pickle

scv.settings.set_figure_params('scvelo')

import matplotlib.colors as mcolors
color_plate = list(mcolors.TABLEAU_COLORS)
```

```python
### option 1: calculate RNA velocity from the beginning.
adata = sc.read("../10_late_blood/exp.mtx", cache=True)
spliced = sc.read("../10_late_blood/exp_exon.mtx", cache=True)
unspliced = sc.read("../10_late_blood/exp_intron.mtx", cache=True)
pdata = pd.read_csv("../10_late_blood/obs.csv", index_col = 0)
fdata = pd.read_csv("../data/cao_beth_var.csv", index_col = 0)

### add spliced and unspliced to layers, pd to obs
adata.layers['spliced'] = spliced.X
adata.layers['unspliced'] = unspliced.X
adata.obs = pdata
adata.var = fdata

adata
scv.utils.show_proportions(adata)
```

```python
print(adata.obs.celltype.value_counts())
print(adata.obs.day.value_counts())
```

```python
print(adata.X.shape)
print(adata.obs)
```

```python
scv.pp.filter_genes(adata, min_counts=5, min_counts_u=5)
scv.pp.normalize_per_cell(adata)
adata.raw = adata
scv.pp.filter_genes_dispersion(adata, n_top_genes=3000)
scv.pp.log1p(adata)
```

```python
scv.pp.moments(adata, n_pcs=30, n_neighbors=30)
scv.tl.velocity(adata)
scv.tl.velocity_graph(adata)
```

```python
#sc.external.pp.bbknn(adata, batch_key='group')
scv.tl.umap(adata, min_dist = 0.75)
```

```python
name = "10_late_blood"

scv.pl.velocity_embedding_stream(adata, 
                                 color = "celltype",
                                 legend_loc='on data',
                                 legend_fontoutline = 2,
                                 figsize = (4,4),
                                 legend_fontsize = 8,
                                 linewidth = 0.5,
                                 sort_order = False,
                                 legend_fontweight = "normal",
                                 title = "",
                                 palette = "Set1",
                                 save = "%s.svg"%name
                                 )
```
